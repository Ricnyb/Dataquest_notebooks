# Pandas barplots

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

# Grouped barplots

### a) Distribution between men and women of top 10 educations

```python
recent_grads[:10].plot.bar(x='Major', y=['Men','Women'])
```

### b) Distribution of men and women in the bottom 10 educations

```python
recent_grads[163:].plot.bar(x='Major', y=['Men','Women'])
```

From the above plots it is obvious that men tend to study more 'hard' sciences while women tend to study more 'soft' sciences.

### c) Bar plot over all majors and median

```python
recent_grads.plot(kind='barh', x='Major', y='Median', figsize=(25,45))
```
